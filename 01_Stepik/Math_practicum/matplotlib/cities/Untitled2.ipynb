```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd

cities = pd.read_csv('cities.csv', sep = ',')
cities = cities[cities['Population'] > 1e6]
cities
```

```python
with open('world.png', 'rb') as pic:
    img = plt.imread(pic)

fig, axes = plt.subplots(figsize=(13, 8))
axes.imshow(img)
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.plot(1280*(cities['Latitude deg'].iloc[0]+180)/360, 644*(-cities['Longitude deg'].iloc[0]+90)/180, 'r*', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[1:3]+180)/360, 644*(-cities['Longitude deg'].iloc[1:3]+90)/180, 'bo', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[3]+180)/360, 644*(-cities['Longitude deg'].iloc[3]+90)/180, 'mo', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[4]+180)/360, 644*(-cities['Longitude deg'].iloc[4]+90)/180, 'g*', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[5]+180)/360, 644*(-cities['Longitude deg'].iloc[5]+90)/180, 'y*', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[6:8]+180)/360, 644*(-cities['Longitude deg'].iloc[6:8]+90)/180, 'yo', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[8]+180)/360, 644*(-cities['Longitude deg'].iloc[8]+90)/180, color='orange', marker='*', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[9]+180)/360, 644*(-cities['Longitude deg'].iloc[9]+90)/180, color='black', marker='o', lw=0.)
axes.plot(1280*(cities['Latitude deg'].iloc[10]+180)/360, 644*(-cities['Longitude deg'].iloc[10]+90)/180, 'w*', lw=0.)
```
