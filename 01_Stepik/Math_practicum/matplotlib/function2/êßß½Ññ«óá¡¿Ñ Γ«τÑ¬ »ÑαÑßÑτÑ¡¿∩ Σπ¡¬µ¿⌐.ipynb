```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

x = np.linspace(-30, 11, 5000)
y = f(x)
fig, axes = plt.subplots(1, 3, figsize=(12,6), dpi=120)
for i in range(3):
    if i < 1:
        x = np.linspace(-30, 1, 500)
        axes[i].set_title(u'f(x) при x $\\to  -\infty$')
        axes[i].set_ylabel('f(x)', fontsize=14)
    elif i > 1:    
        x = np.linspace(-1, 48, 500)
        axes[i].set_title(u'f(x) при x $\\to  +\infty$')
        xsticks = [ 0, 10, 25.8, 41.6]
        axes[i].set_xticks(xsticks)        
        ysticks = [-11.9, 0, 2.5, 20, 40, 57.2]
        axes[i].set_yticks(ysticks)
        axes[i].plot([10, 10], [0, 2.5], '--k', linewidth=0.3) # вертикальная линия        
        axes[i].plot([0, 10], [2.5, 2.5], '--k', linewidth=0.5) # горизонтальная линия  
        axes[i].plot([25.8, 25.8], [0, -11.9], '--k', linewidth=0.5) # вертикальная линия    
        axes[i].plot([0, 25.8], [-11.9, -11.9], '--k', linewidth=0.5) # горизонтальная линия 
        axes[i].plot([41.6, 41.6], [0, 57.2], '--k', linewidth=0.5) # вертикальная линия 
        axes[i].plot([0, 41.6], [57.2, 57.2], '--k', linewidth=0.5) # горизонтальная линия            
    else:
        x = np.linspace(-0.5, 14, 500)
        axes[i].set_title(u'f(x) при небольших x')
        xsticks = [ 0, 4.1,10,20,30]
        axes[i].set_xticks(xsticks)        
        axes[i].set_ylim(1, 5.5)
        ysticks = [1,1.75, 2.5, 5]
        axes[i].set_yticks(ysticks)
        axes[i].plot([4.1, 4.1], [0, 1.75], '--k', linewidth=0.5) # вертикальная линия
        axes[i].plot([0, 4.1], [1.75, 1.75], '--k', linewidth=0.5) # горизонтальная линия  
        axes[i].plot([10, 10], [0, 2.5], '--k', linewidth=0.5) # вертикальная линия        
        axes[i].plot([0, 10], [2.5, 2.5], '--k', linewidth=0.5) # горизонтальная линия         
    y = f(x)
    axes[i].plot(x, y, 'r' )
    axes[i].set_xlabel('x', fontsize=14)
    axes[i].axvline(x=0, color='k', linewidth=0.5)
    axes[i].axhline(y=0, color='k', linewidth=0.5)
plt.suptitle(u'График функции $f(x) = sin(\\frac{x}{5}) \\cdot e^\\frac{X}{10} + 5 \\cdot e^{-\\frac{X}{2}}$')
plt.savefig('fx.svg')
plt.show()
```

При положительных значениях Х функция f(x) носит периодический характер с возрастающей амплитулой при увеличении Х, причем f(x) стремится к $+\infty$.

```python
import numpy as np
import matplotlib.pyplot as plt


def g(x):
    return 1-x**3+x**5


fig, axes = plt.subplots(1, 2, figsize=(11,7), dpi=120)
for i in range(2):
    if i < 1:
        x = np.linspace(-30, 30, 500)
        axes[i].set_title(u'Поведение g(x) при от $-\infty$  до  $+\infty$')
    else:
        x = np.linspace(-1.1, 1.1, 500)
        axes[i].set_title(u'Поведение g(x) при малых x')
        axes[i].set_ylim(0.6, 1.4)
        # пунктир к точке пересечения
        axes[i].plot([-0.78, -0.78], [0, 1.186], '--k', linewidth=0.5) # вертикальная линия
        axes[i].plot([0, -0.78], [1.186, 1.186], '--k', linewidth=0.5) # горизонтальная линия 
#         axes[i].axhline(y=1.185, color='k', linestyle ='--', linewidth=0.5)        
        axes[i].plot([0.78,0.78], [0, 0.814], '--k', linewidth=0.5) # вертикальная линия
        axes[i].plot([0, 0.78], [0.814, 0.814], '--k', linewidth=0.5) # горизонтальная линия 
#         axes[i].axhline(y=0.814, color='k', linestyle ='--', linewidth=0.5)
        xsticks = [-0.78, 0, 0.78]
        axes[i].set_xticks(xsticks)        
        ysticks = [0.814, 1, 1.185]
        axes[i].set_yticks(ysticks)
    y = g(x)
    axes[i].plot(x, y, 'g' )
    axes[i].set_xlabel('x', fontsize=10)
    axes[i].set_ylabel('g(x)', fontsize=10)
    axes[i].axvline(x=0, color='k', linewidth=0.5)
    axes[i].axhline(y=0, color='k', linewidth=0.5)
    axes[i].axhline(y=1, color='k', linewidth=0.5, linestyle ='--')
plt.suptitle(u'График функции $g(x) = 1 - x^3 + x^5$')
plt.savefig('gx.svg')
plt.show()
```

Значение функции g(x) возрастает и стремится к $+\infty$, при при увеличении Х до $+\infty$.
Значение функции g(x) убывает и стремится к $-\infty$, при при стремлении Х к $-\infty$.
При малых значениях Х (на промежутке х=(-1,1) функция g(x) имеет точку перегиба при х=0, локальный максимум при х-0.78 и локальный минимум при х=0.78.

```python
import numpy as np
import matplotlib.pyplot as plt


def h(x):
    return (5*np.sin(np.exp(0.8*x))**2 * x**2) / np.exp(x**2/2)


fig, axes = plt.subplots(1, 2, figsize=(12,8), dpi=120)
for i in range(2):
    ysticks = [0, 0.6, 0, 2.13, 2.91, 1.23, 0.51, 0.23, 0.11]
    if i < 1:
        x = np.linspace(-11, 11, 5000)
        axes[i].set_title(u'Поведение h(x) при от $-\infty$  до  $+\infty$')
    else:
        x = np.linspace(-3, 3.7, 5000)
        axes[i].set_title(u'Поведение h(x) при малых x')
        xsticks = [-3, -0.86, 0, 0.86, 1.91, 2.55]  # [2.98, 3.3, 3.55]
        # пунктир к точке пересечения
        for xs in range(len(xsticks)):
            axes[i].plot([xsticks[xs], xsticks[xs]], [0, ysticks[xs]], '--k', linewidth=0.5) # вертикальная линия        
        xsticks0 = [0, 1.43, 2.3, 2.8, 3.16, 3.44, 3,67, 3.86]
        axes[i].set_xticks(xsticks+[3.55])        
  
    y = h(x)
    axes[i].set_yticks(ysticks)       
    for ys in ysticks:
        axes[i].axhline(y=ys, color='k', linestyle ='--', linewidth=0.3)   
    axes[i].plot(x, y, 'b' )
    axes[i].set_xlabel('x', fontsize=14)
    axes[i].set_ylabel('h(x)', fontsize=14)
    axes[i].axvline(x=0, color='k', linewidth=0.5)
    axes[i].axhline(y=0, color='k', linewidth=0.5)
plt.suptitle(u'График функции $h(x) = \\frac{5 \\cdot sin^2(e^{0.8 \\cdot x}) \\cdot x^2}{e^{\\frac{x^2}{2}}}$')
plt.savefig('hx.svg')
plt.show()
```

Значение функции h(x) стремится к 0 при при стремлении Х к $-\infty$ или к $+\infty$.
При значении Х=-4 функция g(x) начинает приобретать периодический характер синусоидального вида с уменьшением периода колебаний при увеличении Х, достигающая пика при Х=1.9 и затем амплитуда колебаний спадает почти до нуля при Х=6. Это хорошо видно на графиках в следующей секции.

```python
import numpy as np
import matplotlib.pyplot as plt

def h(x):
    return (5*np.sin(np.exp(0.8*x))**2 * x**2) / np.exp(x**2/2)

fig, axes = plt.subplots(1, 2, figsize=(12,6), dpi=120)
for i in range(2):
    if i < 1:
        x = np.linspace(-5, -4, 5000)
        axes[i].set_title(u'Поведение h(x) при x стремящемся к $-\infty$')
    else:
        x = np.linspace(5, 6, 5000)
        axes[i].set_title(u'Поведение h(x) при x стремящемся к  $+\infty$')
    y = h(x)
    axes[i].plot(x, y, 'b' )
    axes[i].set_xlabel('x', fontsize=14)
    axes[i].set_ylabel('h(x)', fontsize=12)
    axes[i].axhline(y=0, color='k', linewidth=0.5)
plt.suptitle(u'График функции $h(x) = \\frac{5 \\cdot sin^2(e^{0.8 \\cdot x}) \\cdot x^2}{e^{\\frac{x^2}{2}}}$')
plt.savefig('hx-dtl.svg')
plt.show()
```

Значение функции h(x) стремится к 0 при при стремлении Х к $-\infty$ или к $+\infty$, причем в сторону отрицательных Х - функция h(x) апериаодическая, а в сторону положительных Х - периодическая с уменьшающейся амплитудой.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

def g(x):
    return 1-x**3+x**5

def h(x):
    return (5*np.sin(np.exp(0.8*x))**2 * x**2) / np.exp(x**2/2)

cg = ['r','g','b']
fig, axes = plt.subplots(1, 2, figsize=(12,7), dpi=120)
for i in range(2):
    if i < 1:
        x = np.linspace(-7, 63.2, 10000)
        axes[i].set_title(u'Поведение f(x), g(x), h(x) при больших x')
        axes[i].set_ylim(min(f(x)-2), max(f(x)+2))
    else:
        x = np.linspace(-1.3, 2.3, 10000)
        axes[i].set_title(u'Поведение f(x), g(x), h(x) при малых x')
        axes[i].set_ylim(-0.2, 3.5)
    for n in range(3):
        if n == 1:
            y = g(x)
            gx = y
            nf = 'g(x)'
        elif n == 2:
            y = h(x)
            hx = y
            nf = 'h(x)'
        else:
            y = f(x)
            fx = y
            nf = 'f(x)'
        axes[i].plot(x, y, cg[n], label=nf)
    axes[i].set_xlabel('x', fontsize=14)
    axes[i].set_ylabel('f(x), g(x), h(x)', fontsize=10)
    axes[i].axvline(x=0, color='k', linewidth=0.5)
    axes[i].axhline(y=0, color='k', linewidth=0.5)
    axes[i].legend(loc=i+1)
    if i:
        xysticks = []
        idx = np.argwhere(np.diff(np.sign(fx - gx))).flatten()
        for ind in idx: xysticks.append([x[ind], fx[ind], gx[ind]])
        idx = np.argwhere(np.diff(np.sign(fx - hx))).flatten()
        for ind in idx: xysticks.append([x[ind], fx[ind], hx[ind]])
        idx = np.argwhere(np.diff(np.sign(gx - hx))).flatten()
        for ind in idx: xysticks.append([x[ind], gx[ind], hx[ind]])
        xsticks = [round(xy[0],2) for xy in xysticks]
        ysticks = [round(xy[1],2) for xy in xysticks]
        # пунктир к точке пересечения
        for xs in range(len(xsticks)):
            axes[i].plot([xsticks[xs], xsticks[xs]], [0, ysticks[xs]], '--k', linewidth=0.6) # вертикальная линия
            axes[i].plot([0, xsticks[xs]], [ysticks[xs], ysticks[xs]], '--k', linewidth=0.6)  # горизонтальная линия
#             print(f'({xsticks[xs],ysticks[xs]})')
        xsticks = xsticks + [0]
        ysticks = ysticks + [0]
        xlabels = [str(xx) for xx in xsticks]
        axes[i].set_xticks(xsticks)
        axes[i].set_xticklabels(xlabels, rotation=90, fontsize=8)
        axes[i].set_yticks(ysticks)
    else:
        xsticks = [-7, 0, 5, 15.7, 31.4, 47.1]
        axes[i].set_xticks(xsticks)
plt.suptitle(u'Графики функций $f(x) = sin(\\frac{x}{5}) \\cdot e^\\frac{X}{10} + 5 \\cdot e^{-\\frac{X}{2}}$,  $g(x) = 1 - x^3 + x^5$,  $h(x) = \\frac{5 \\cdot sin^2(e^{0.8 \\cdot x}) \\cdot x^2}{e^{\\frac{x^2}{2}}}$')
plt.savefig('fgh.svg')
plt.show()
```
