# Исследование точек пересечения функций
### Задание 1

$$f(x) = sin(\frac{x}{5}) \cdot e^\frac{x}{10} + 5\cdot e^{-\frac{x}{2}}$$, 

$$g(x)= 1-x^3+x^5$$, 

$$h(x)=\frac{5\cdot sin^2(e^{0.8x})\cdot x^2}{e^\frac{x^2}{2}}$$

Каждая группа должна демонстрировать:
- общее поведение функции (на −∞ -\infty −∞ и +∞ +\infty +∞)
- периодичность или апериодичность (на всей области определения или на отдельных участках)
- экстремумы
- перегибы

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2) 

def g(x):
    return 1 - x**3 + x**5

def h(x):
    return 5 * (np.sin(np.exp(0.8 * x)))**2 * x**2 / (np.exp(x**2/2))
```

```python
x = np.arange(-50, 400, 0.1)
y = [f(i) for i in x]

fig, axes = plt.subplots(1, 3, figsize=(13,8), dpi=120)
plt.suptitle(r'Графики функции $f(x) = sin(\frac{x}{5}) \cdot e^\frac{x}{10} + 5\cdot e^{-\frac{x}{2}}$', fontsize=12)

axes[0].grid()
axes[0].set_ylim([-100,100])
axes[0].set_xlim([-20, 50])
axes[0].plot(x, y, 'r', label='Функция $f(x)$')
axes[0].legend(loc=0, fontsize=12)
axes[0].set_xlabel('x',fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)

axes[1].grid()
axes[1].set_ylim([-10**8,10**8])
axes[1].set_xlim([-50, 200])
axes[1].plot(x, y, 'r', label='Функция $f(x)$')
axes[1].legend(loc=0, fontsize=12)
axes[1].set_xlabel('x',fontsize=12)
axes[1].set_ylabel('f(x)', fontsize=12)

axes[2].grid()
axes[2].set_ylim([-10**13,10**13])
axes[2].set_xlim([200, 400])
axes[2].plot(x, y, 'r', label='Функция $f(x)$')
axes[2].legend(loc=0, fontsize=12)
axes[2].set_xlabel('x',fontsize=12)
axes[2].set_ylabel('f(x)', fontsize=12)
```

Выводы:
- функция стремится в бесконечность при отрицательных X
- при положительных Х функция начинает совершать экспоненциально нарастающие периодическоие колебания
- в виду такого поведения графиков поиск экстремумов рационально совершать на заранее определенном участке


```python
x = np.arange(-100, 100, 0.1)
y = [g(i) for i in x]
fig, axes = plt.subplots(1, 2, figsize=(13,8), dpi=120)
plt.suptitle(r'Графики функции $g(x)= 1-x^3+x^5$', fontsize=12)

axes[0].grid()
# axes[0].set_ylim([-100,100])
# axes[0].set_xlim([-20, 50])
axes[0].plot(x, y, 'r', label='Функция $g(x)$')
axes[0].legend(loc=0, fontsize=12)
axes[0].set_xlabel('x',fontsize=12)
axes[0].set_ylabel('g(x)', fontsize=12)

axes[1].grid()
axes[1].set_ylim([0,1.5])
axes[1].set_xlim([-2, 2])
axes[1].plot(x, y, 'r', label='Функция $g(x)$')
axes[1].legend(loc=0, fontsize=12)
axes[1].set_xlabel('x',fontsize=12)
axes[1].set_ylabel('g(x)', fontsize=12)
```

Выводы:
- функция c быстрым ростом в бесконечность с ростом х
- на промежутке от -1 до 1 имеет два экстремума
- в точке (1,0) имеет перегиб
- функция не является периодичной

```python
x = np.arange(-20, 20, 0.1)
y = [h(i) for i in x]
fig, axes = plt.subplots(1, 2, figsize=(13,8), dpi=120)
plt.suptitle(r'Графики функции $h(x)=\frac{5\cdot sin^2(e^{0.8x})\cdot x^2}{e^\frac{x^2}{2}}$', fontsize=12)

axes[0].grid()
# axes[0].set_ylim([-0.1,0.1])
axes[0].set_xlim([-20, 20])
axes[0].plot(x, y, 'r', label='Функция $h(x)$')
axes[0].legend(loc=0, fontsize=12)
axes[0].set_xlabel('x',fontsize=12)
axes[0].set_ylabel('h(x)', fontsize=12)

axes[1].grid()
axes[1].set_ylim([-0.1,3])
axes[1].set_xlim([-3.5, 5.5])
axes[1].plot(x, y, 'r', label='Функция $h(x)$')
axes[1].legend(loc=0, fontsize=12)
axes[1].set_xlabel('x',fontsize=12)
axes[1].set_ylabel('h(x)', fontsize=12)
```

Выводы:
- функция стремится к нулю при росте |x| 
- на промежутке [-3; 5] функция имеет 10 выраженных максимумов и 9 минимумов
- функция является периодичной

### Задание 2

```python
x = np.arange(-2, 5, 0.01)


fig = plt.figure(figsize=(13, 8), dpi=120)
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim([-5,5])
axes.set_xlim([-2, 5])
axes.grid()

yr = [f(i) for i in x]
axes.plot(x, yr, 'r', label='Функция $f(x$') 
gr = [g(i) for i in x]
axes.plot(x, gr, 'g', label='Функция $g(x$') 
hr = [h(i) for i in x]
axes.plot(x, hr, 'b', label='Функция $h(x$') 

axes.legend(loc=4, fontsize=12)
axes.set_xlabel('x',fontsize=12)
axes.set_ylabel('f(x), g(x), h(x)', fontsize=12)
axes.set_title(r'Графики функции $f(x), g(x), h(x)$', fontsize=12)
fig.savefig('outgraph.svg')
x_cross = []
y_cross = []

for i in np.arange(-1.5, 3, 0.01):
    if abs(g(i) - h(i)) < 0.02 or abs(f(i) - g(i)) < 0.06:
        x_cross.append(round(i, 2))
        y_cross.append(round(g(i), 2))
    if abs(f(i) - h(i)) < 0.04:
        x_cross.append(round(i, 2))
        y_cross.append(round(f(i), 2))
    
for i in range(len(x_cross)):
    plt.scatter(x_cross[i], y_cross[i], color='k', marker='X', s=150)

x = np.arange(-2, 35, 0.1)
yr = [f(i) for i in x]
gr = [g(i) for i in x]
hr = [h(i) for i in x]
axes2 = fig.add_axes([0.15, 0.06, 0.7, 0.4])
axes2.set_ylim([-12,10])
axes2.set_xlim([-2, 35])
axes2.grid()

axes2.plot(x, yr, 'r', label='Функция $f(x$') 
axes2.plot(x, gr, 'g', label='Функция $g(x$') 
axes2.plot(x, hr, 'b', label='Функция $h(x$') 
axes2.legend(loc=3, fontsize=12)
axes2.set_xlabel('x',fontsize=12)
axes2.set_ylabel('f(x), g(x), h(x)', fontsize=12)
axes2.set_title(r'Графики функции $f(x), g(x), h(x)$ при больших х', fontsize=12)
axes2.plot()

dots = (dict(zip(x_cross, y_cross)))
print(dots)
```

Выводы:
- на промежутке малых значений х функции имеют изображенные на графике 6 точек перечечения
- значения точек пересечения, полученные машинным способом и округленные до 2 знаков после запятой сохранены в словарь dots
- при больших значениях х, вункция f(x), в силу своей периодичности, будет многократно пересекать функцию h(x)
