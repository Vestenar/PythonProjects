```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5.) * np.exp(x / 10.) + 5. * np.exp(-x / 2)

def g(x):
    return 1 - x * x * x + x * x * x * x * x

def h(x):
    return 5 * (x * np.sin(np.exp(0.8 * x))) ** 2 / np.exp((x ** 2) / 2.)
```

```python
x = np.arange(-100, 100, 0.1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 8))
plt.suptitle(r'График функции $𝑓(𝑥)=\sin(\frac{𝑥}{5})\cdot e^\frac{x}{10}+5\cdot e^{-\frac{x}{2}}$', fontsize=18)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title(r'при $x \to -\infty$', fontsize=12)
axes[0].plot(x-1000, f(x-1000), 'r')
axes[0].set_yscale('log')

axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('f(x)', fontsize=12)
axes[1].set_title('в окрестности нуля', fontsize=12)
axes[1].plot(x, f(x), 'r')
axes[1].set_ylim(-10000, 10000)
axes[1].set_xlim(-20, 100)

axes[2].set_xlabel('x', fontsize=12)
axes[2].set_ylabel('f(x)', fontsize=12)
axes[2].set_title(r'при $x \to +\infty$', fontsize=12)
axes[2].plot(x+1000, f(x+1000), 'r', label='$f(x)$')
axes[2].plot(x+1000, -f(x+1000), 'r--', label='$-f(x)$')
axes[2].set_yscale('log')
axes[2].legend(loc=2, fontsize=12)

plt.show()
```

ВЫВОДЫ:

Функция несимметрична. При положительных x, дает больший вклад член $\sin(\frac{𝑥}{5})\cdot e^\frac{x}{10}$ и график представляет собой периодическую функцию с экспоненциально нарастающей амплитудой. При отрицательных x, быстро нарастает второй член $5\cdot e^{-\frac{x}{2}}$ и график демонстрирует экспоненциальный рост при больших отрицательных x. Из-за колебательного характера, имеет бесконечное число экстремумов и точек перегиба (все при пположительных x).

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 8))
plt.suptitle(r'График функции $g(𝑥)=1-x^3+x^5$', fontsize=18)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('g(x)', fontsize=12)
axes[0].set_title(r'при $x \to -\infty$', fontsize=12)
axes[0].plot(x-1000, g(x-1000), 'g')

axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('g(x)', fontsize=12)
axes[1].set_title('в окрестности нуля', fontsize=12)
axes[1].plot(x/10, g(x/10), 'g')
axes[1].set_ylim(0, 2)
axes[1].set_xlim(-1.5, 1.5)

axes[2].set_xlabel('x', fontsize=12)
axes[2].set_ylabel('g(x)', fontsize=12)
axes[2].set_title(r'при $x \to +\infty$', fontsize=12)
axes[2].plot(x + 1000, g(x + 1000), 'g')

plt.show()
```

ВЫВОДЫ:

Функция g(x) - антисимметрична относительно точки (0,1).
При  очень большом положительном или отрицательном аргументе - ведет себя как функция $x^5$.
Имеет 2 екстремума и 3 точкиперегиба (все в небольшой ([-1, 1]) окрестности нуля).

```python
fig, axes = plt.subplots(figsize=(15, 8))
plt.suptitle(r'График функции $h(𝑥)=\frac{5\cdot\sin^2(e^{0.8\cdot x})\cdot x^2}{e^\frac{x^2}{2}}$', fontsize=18)

axes.set_xlabel('x', fontsize=12)
axes.set_ylabel('h(x)', fontsize=12)
axes.set_title('в окрестности нуля', fontsize=12)
axes.plot(x/10., h(x/10.), 'b', label='$h(x)$')
axes.set_ylim(-0.2, 3)
axes.legend(loc=2, fontsize=18)

axes2 = axes.twinx()
axes2.plot(x/10., h(x/10.), 'y:', label='$h(x)$')
#axes2.set_xlim(-2, 5)
axes2.set_yscale('log')
axes2.legend(loc=1, fontsize=18)

plt.show()
```

ВЫВОДЫ:

Положительно определенная функция на всей оси. При положительных x, имеет периодический характер, с увеличивающейся по мере роста х частотой. Огибающая при больших по модулю х имеет характер $e^\frac{-x^2}{2}$
