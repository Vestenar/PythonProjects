```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return math.sin(x / 5) * math.e ** (x / 10) + 5 * math.e ** (-x / 2)
def g(x):
    return 1 - x ** 3 + x ** 5
def h(x):
    return (5 * math.sin(math.e ** (0.8 * x)) ** 2 * (x ** 2)) / (math.e ** ((x ** 2) / 2))
```

# Задание 1

```python
colors = ['blue', 'red', 'green']
titles = ['f(x)', 'g(x)', 'h(x)']
fig, axes = plt.subplots(3, 1)
fig.set_size_inches((10, 17))
x = np.arange(-10, 10, 0.1)
ys = []
for i in range(3):
    t = []
    for p in x:
        if i == 0:
            y = f(p)
        elif i == 1:
            y = g(p)
        else:
            y = h(p)
        t.append(y)
    ys.append(t)
for i in range(len(ys)):
    axes[i].plot(x, ys[i], color=colors[i])
    axes[i].set_xlabel('x')
    axes[i].set_ylabel('y')
    axes[i].set_title(titles[i])
fig.suptitle('При больших значениях')
```

# Задание 2

```python
fig = plt.figure(figsize=(10, 10))
axes = fig.add_axes([0, 0, 1, 1])
x = np.arange(-2, 5, 0.001)
ys = []
for i in range(3):
    t = []
    for p in x:
        if i == 0:
            y = f(p)
        elif i == 1:
            y = g(p)
        else:
            y = h(p)
        t.append(y)
    ys.append(t)
for i in range(len(ys)):
    axes.plot(x, ys[i], label=titles[i], color=colors[i])
    axes.set_xlabel('x')
    axes.set_ylabel('y')
axes.set_ylim((-2, 5))
axes.grid()
fig.suptitle('Точки пересечения', fontsize=16)
for i in range(len(x)):
    if abs(ys[0][i] - ys[1][i]) < 0.01:
        axes.scatter(x[i], ys[0][i], color='cyan', zorder=100)
    if abs(ys[0][i] - ys[2][i]) < 0.01:
        axes.scatter(x[i], ys[0][i], color='orange', zorder=100)
    if abs(ys[1][i] - ys[2][i]) < 0.01:
        axes.scatter(x[i], ys[2][i], color='black', zorder=100)
fig.legend()
fig.show()
```
