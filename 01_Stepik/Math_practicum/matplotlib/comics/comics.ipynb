```python
import matplotlib.pyplot as plt

def edit_image(ax, image, extent=[0,3,0,4]):
    ax.imshow(image, extent=extent)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)


with open('1.png', 'rb') as map_file:
    map_img1 = plt.imread(map_file)
with open('2.png', 'rb') as map_file:
    map_img2 = plt.imread(map_file)
with open('3.png', 'rb') as map_file:
    map_img3 = plt.imread(map_file)
with open('4.png', 'rb') as map_file:
    map_img4 = plt.imread(map_file)


fig, axes = plt.subplots(2, figsize=(8, 8))

edit_image(axes[0], map_img1, [0,6,0,3])
edit_image(plt.subplot(234), map_img2)
edit_image(plt.subplot(235), map_img3)
edit_image(plt.subplot(236), map_img4)

plt.tight_layout()
plt.savefig("5.png")
plt.show()

```
