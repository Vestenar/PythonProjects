```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python

fig = plt.figure()

# Сетка состоит из 2 строк и 3 столбцов. Первая ячейка имеет номер (0,0),
egrid = (2,3)
ax1 = plt.subplot2grid(egrid, (0, 0), colspan=3)
ax2 = plt.subplot2grid(egrid, (1, 0), rowspan=1)
ax3 = plt.subplot2grid(egrid, (1, 1), rowspan=1)
ax4 = plt.subplot2grid(egrid, (1, 2), rowspan=1)

#fig, axes = plt.subplots(1, 3)


ax1.get_xaxis().set_visible(False)
ax1.get_yaxis().set_visible(False)
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)
ax3.get_xaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)
ax4.get_xaxis().set_visible(False)
ax4.get_yaxis().set_visible(False)


img1 = open('1.png','rb')
img2 = open('2.png','rb')
img3 = open('3.png','rb')
img4 = open('4.png','rb')
map_ing1 = plt.imread(img1)
map_ing2 = plt.imread(img2)
map_ing3 = plt.imread(img3)
map_ing4 = plt.imread(img4)

ax1.imshow(map_ing1, aspect='auto')
ax2.imshow(map_ing2, aspect='auto')
ax3.imshow(map_ing3, aspect='auto')
ax4.imshow(map_ing4, aspect='auto')

#fig.tight_layout()

plt.show()
fig.savefig('test.png')
```
