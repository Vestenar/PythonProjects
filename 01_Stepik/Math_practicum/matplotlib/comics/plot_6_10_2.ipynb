```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(figsize=(13,8))

with open('1.png', 'rb') as map_file:
    map_img = plt.imread(map_file)

axes.imshow(map_img)
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)

fig, axes = plt.subplots(1, 3, figsize=(13,8))
for i in range(2,5):
    with open('{}.png'.format(i), 'rb') as map_file:
        map_img = plt.imread(map_file)
    axes[i-2].imshow(map_img)
    axes[i-2].get_xaxis().set_visible(False)
    axes[i-2].get_yaxis().set_visible(False)
plt.show()


```
