```python
import matplotlib.pyplot as plt

from matplotlib import rcParams

%matplotlib inline
```

```python
def plot_image(ax, img_numb):
    with open('{}.png'.format(img_numb), 'rb') as inp:
        img = plt.imread(inp)
        ax.imshow(img, aspect="auto")
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)


rcParams["image.aspect"] = 'equal'
fig = plt.figure()

ax1 = plt.subplot(211)
fig.set_size_inches(8, 9)
ax2 = plt.subplot(234)
ax3 = plt.subplot(235)
ax4 = plt.subplot(236)

plot_image(ax1, 1)
plot_image(ax2, 2)
plot_image(ax3, 3)
plot_image(ax4, 4)

plt.tight_layout()
```
