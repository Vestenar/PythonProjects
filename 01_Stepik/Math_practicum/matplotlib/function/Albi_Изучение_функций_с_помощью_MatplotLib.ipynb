## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
# Исследуемая функция
def func(a,b,x):
    return ((x**b)+(a**b))/(x**b)
 # наши коэффиценты   
ab_values = ((1, 1, "r"),
            (2, 1, "g"),
            (1, 2, "b")
                )
```

```python
#Построить график (размер графика должен быть достаточным, чтобы визуально увидеть особенности изучаемых функций), 
#график каждой функции должен быть одного цвета для одного значения  α  и  β .

x1_1 = np.arange(-3,-0.0001, 0.0001)
y1_1 = func(1,1,x1_1)
x1_2 = np.arange(0.0001,3, 0.0001)
y1_2 = func(1,1,x1_2)

# alpha = 1, beta = 1 
plt.figure(figsize=(13,8)) # поле для отображения
plt.title('Часть 1. График функции $ \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta}}$ для различных $\\alpha$ и $\\beta $')#Подписать заголовок
plt.plot(x1_1, y1_1, 'r', label='$ \\alpha = 1, \\beta = 1 $')
plt.plot(x1_2, y1_2, 'r')
plt.xlabel(u'x')#Подписать оси 
plt.ylabel(u'f(x)')
plt.ylim(top=50, bottom=-50) #ограничения для наглядности
plt.xlim(-3, 3)

# alpha = 2, beta = 1 
x2_1 = np.arange(-3,-0.0001, 0.0001)
y2_1 = func(2,1,x2_1)
x2_2 = np.arange(0.0001,3, 0.0001)
y2_2 = func(2,1,x2_2)
plt.plot(x2_1, y2_1, 'g', label='$ \\alpha = 2, \\beta = 1 $')
plt.plot(x2_2, y2_2, 'g')

# alpha = 1, beta = 2 
x3_1 = np.arange(-3,-0.0001, 0.0001)
y3_1 = func(1,2,x3_1)
x3_2 = np.arange(0.0001,3, 0.0001)
y3_2 = func(1,2,x3_2)
plt.plot(x3_1, y3_1, 'm' , label='$ \\alpha = 1, \\beta = 2 $')
plt.plot(x3_2, y3_2, 'm')
plt.savefig("6_10_Matplotlib-1.svg", format="svg") #Сохранить изображение в svg файл
plt.legend()#Создать легенду
```

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
fig = plt.figure(figsize=(13, 8))

x = np.arange(0.01, 5, 0.01)
for a, b, color in ab_values:
    plt.plot(x, func(a,b,x), color, label=f"$ \\alpha = {a}, \\beta = {b} $")

plt.ylim(0, 40)
plt.legend()#Создать легенду
title = "Часть 2. График функции $ \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta}}$ для различных $\\alpha$ и $\\beta $ при х > 0"

axes_0 = fig.add_axes([0.25, 0.47, 0.20, 0.35])#добавляем оси для врезок
x_0 = np.arange(0.001, 1.5, 0.001)
axes_1 = fig.add_axes([0.5, 0.47, 0.3, 0.35])
x_1 = np.arange(2, 40, 0.5)

for axe, x in ((axes_0, x_0), (axes_1, x_1)):
    for a, b, color in ab_values:
        axe.plot(x, func(a,b,x), color)
    axe.set_xlabel("x")
    axe.set_ylabel("f(x)")
axes_0.set_title("График функций для малых х")
axes_1.set_title("График функций для больших х")

axes_0.set_ylim(0, 15)#ограничения для наглядности
plt.savefig("6_10_Matplotlib-2.svg", format="svg") #Сохранить изображение в svg файл
```

### Часть 3

Построить в общих осях графики для $x<0$.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
fig = plt.figure(figsize=(15, 10))

x = np.arange(-5, 0.01, 0.01)
for a, b, color in ab_values:
    plt.plot(x, func(a,b,x), color, label=f"$ \\alpha = {a}, \\beta = {b} $")
plt.ylim(-30, 30)
title = "Часть 3. График функции $ \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta}}$ для различных $\\alpha$ и $\\beta $ при х < 0"
plt.plot(x,1-func( 0, 1,x), 'y', label="f(x)=0")#f(x)=0 получаем из уравнения функции (асимптота минус 1)
plt.legend()#Создать легенду

axes_1 = fig.add_axes([0.25, 0.57, 0.3, 0.25])
x_1 = np.arange(-0.5, -51, -0.5)

for a, b, color in ab_values:
    axes_1.plot(x_1, func(a,b,x_1), color)

axes_1.plot(x_1, 1 - func( 0, 1, x_1), 'y')

axes_1.set_title("График функций при удалении x от 0 к  −∞")

axes_1.set_ylim(-2, 2)
axes_1.set_xlim(-30, 0)

axes_1.set_xlabel("x")
axes_1.set_ylabel("f(x)")

axes_1.set_yticklabels(['{}'.format(y) for y in np.arange(-2, 3, 0.5)])
axes_1.set_xticklabels(['{}'.format(y) for y in np.arange(-30, 5, 5)])

plt.savefig("6_10_Matplotlib-3.svg", format="svg") #Сохранить изображение в svg файл
```

## Задача №2

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

Сделайте выводы о поведении графиков

```python
# Исследуемая функция
def func(a,b,x):
    return ((x**b)+(a**b))/(x**b)
 # наши коэффиценты   
ab_values_2 = ((1, 0, "r"),
            (0, 0, "g"),
            (0, 1, "b"),
            (0, 2, "c"),
            (1, -1, "m"),
            (2, -1, "y")
                )
```

```python
fig = plt.figure(figsize=(12, 8))
x = np.arange(-5, 5, 0.01)
for a, b, color in ab_values_2:
    plt.plot(x, func(a,b,x), color, label=f"$ \\alpha = {a}, \\beta = {b} $")
plt.ylim(-3, 3)
plt.legend()#Создать легенду
plt.savefig("6_10_Matplotlib-4.svg", format="svg") #Сохранить изображение в svg файл
```

1. Все функции представляют собой прямые.
2.  $\alpha = 0$ а $\beta$ равна константе. В этом случае общая функция сводится к виду $1 + 0^{\beta} = 1$, что мы можем наблюдать на 3 и 4 функциях.
3. Если $\alpha \ne 0$ а $\beta = 0$. В этом случае функция сводится к виду $1 + \left( \frac{\alpha}{x} \right)^{0} = 1 + 1 = 2$, что мы наблюдаем для функции 1.
4. $\alpha = \beta = 0$, в данном случае получается неопределенность вида $0^0$ , но принято что 0^0 = 1, поэтому f(x) = 1 + 1 = 2 на всей области х.
5. Если $\beta = -1$ а $\alpha \ne 0$ -  график прямой $ 1 + \frac{x}{\alpha}$, где $1/\alpha$ является тангенсом наклона прямой, а 1 задает смещение относительно (0, 0).

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
# Исследуемая функция
def func(a,b,x):
    return ((x**b)+(a**b))/(x**b)
 # наши коэффиценты   
ab_values_3 = ((1, 0.5, "r"),
            (1, -0.5, "g"),
            (1, -1.5, "b"),
                )
fig = plt.figure(figsize=(10, 8))
x = np.arange(0.0001, 5, 0.01)
for a, b, color in ab_values_3:
    plt.plot(x, func(a,b,x), color, label=f"$ \\alpha = {a}, \\beta = {b} $")
plt.ylim(0, 5)
plt.xlim(-0.05, 5)
plt.title ("Задание 3. График функции $ \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta}}$ для различных $\\alpha$ и $\\beta $")
plt.legend()#Создать легенду
plt.savefig("6_10_Matplotlib-5.svg", format="svg") #Сохранить изображение в svg файл
```

1. Все графики пересекаются в точке (1, 2)
2. Функции монотонны, 1я убывает, 2-3 возрастают.
3. Функция определена только при x > 0. 
4. 2 я функция выпукла вверх, 1я и 3 вогнута (выпуклы вниз). По определению:  Выпуклая функция (выпуклая вниз функция) — функция, для которой любой отрезок между двумя любыми точками графика функции в векторном пространстве лежит не ниже соответствующей дуги графика. 
Вогнутая функция (выпуклая вверх функция) — функция, хорда между двумя точками графика которой лежит не выше образованной дуги графика.
5. Ассимптотика. Первая функция ($ \alpha = 1, \space \beta = 0.5$) на бесконечности стремиться к единице, В нуле функция стемиться к $+ \infty$
Остальные функции стремятся к $+\infty$, т.к. степень при х положительна и больше 0. В нуле функции существуют и равны 1.

## Задача №4

В результате выполнения предыдущей задачи, вы вероятно заметите, что все графики с $\alpha=1$ проходят через общую точку (1, 2).

Постройте в одном ряду 3 графика, чтобы убедиться в выводах, сделанных по результатам предыдущей задачи.

Каждый график будет содержать 4 кривые. 2 общих:

- $\alpha = 1, \beta = 0$ (в качестве цвета попробуйте использовать 'b--')

- $\alpha = 1, \beta = -1$ (в качестве цвета попробуйте использовать 'r--')

И 2 уникальных для каждого графика:

1. $\alpha = 1, \beta = 0.5$ и $\alpha = 1, \beta = 0.8$

2. $\alpha = 1, \beta = -0.5$ и $\alpha = 1, \beta = -0.8$

3. $\alpha = 1, \beta = -1.5$ и $\alpha = 1, \beta = -2.5$

Не забудьте добавить легенду на каждый график. Для этого может потребоваться вызвать метод legend() для каждого объекта осей.

Для того чтобы задать общий заголовок для всех 3 графиков используйте метод https://matplotlib.org/api/_as_gen/matplotlib.pyplot.suptitle.html

```python
# Исследуемая функция
def func(a,b,x):
    return ((x**b)+(a**b))/(x**b)
  
fig, axes = plt.subplots(1, 3, figsize=(14,8))

plt.subplots_adjust(wspace=0.3)
 # наши коэффиценты 
ab_values_4_1 = [
    (1, 0, "b--"),
    (1, -1, "r--")
    ]

ab_values_4_2 = [
    [(1, 0.5, "#DD16D1"), (1, 0.8, "#E8AC3A")],
    [(1, -0.5, "#DD163A"), (1, -0.8, "#A4E522")],
    [(1, -1.5, "#22E9C8"), (1, -2.5, "#9D97BC")],
    ]

x = np.arange(0.001, 6, 0.01)

for i in range(3):
    for a, b, color in ab_values_4_1 + ab_values_4_2[i]:
        axes[i].plot(x, func(a,b,x), color, label= "$\\alpha = {0}, \\beta = {1}$".format(a, b))
        
        axes[i].set_ylim(0.75, 4.5)
        axes[i].set_xlim(-0.1, 4)
        
        axes[i].set_title('Группа {}'.format(i))        
        axes[i].set_xticklabels(['{}'.format(y) for y in np.arange(-1, 5, 1)])
        axes[i].set_yticklabels(['{}'.format(y) for y in np.arange(0.5, 5, 0.5)])        
        axes[i].set_xlabel("x")
        axes[i].set_ylabel("f(x)")  
        axes[i].legend()     

fig.suptitle('Задание 4. График функции $ \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta}}$ для различных $\\alpha$ и $\\beta $')
fig.savefig("6_10_Matplotlib-6.svg")
```

1. Графики функций с  $\alpha$ = 1 пересекаются в точке (1, 2)
2. У не общих функций первой группы показатели у х меньше 0 => функции стремятся к 1 при x -> $+ \infty $, 
3. у функций других групп показатели при х больше 0, поэтому они стремятся к $+ \infty$.
