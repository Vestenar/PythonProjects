# Изучение функций с помощью MatplotLib
Цель задания - исследование функции $$f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$$
при различных значениях $\alpha$ и $\beta$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Определение функции
def f(x, a, b):
  return (x**b + a**b) / x**b
```

## Задача №1
### Часть 1

```python
x = np.arange(-20, 20, 0.01)
task = [(1, 1, 'r'), (2, 1, 'g'), (1, 2, 'b')]  # определяем параметры графика и порядок цвета
fig = plt.figure(figsize=(13, 8), dpi=120)
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim([-20,20])
axes.grid()
for a, b, c in task:
  y = [f(i, a, b)  if abs(i) > 0.05 else np.nan for i in x]
  axes.plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$') 

axes.legend(loc=4, fontsize=12)
axes.set_xlabel('x',fontsize=12)
axes.set_ylabel('f(x)', fontsize=12)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных значениях $\alpha$ и $\beta$', fontsize=12)
fig.savefig('result.svg')
```

Выводы:
- как и следовало ожидать, функция является асимптотической
- при четных значениях $\alpha$ функция является четной, график не пересекает ось иксов
- наглядно демонстрируется невозможность существования функции в четвертой четверти (x > 0, y < 0)

### Часть 2

- Для демонстрации возможности пересечения была добавлена третья врезка, увеличивающая область, где синий график пересекает другие два
- Очевидно, что при малых значениях х рост первого графика быстрее, чем второго, поэтому они не пересекутся даже при предельно малом х
- На графиках наглядно прослеживается асимптотическое стремление исследуемой функции к 1 при стремлении х в бесконечность и к оси ординат при предельно малых значениях х

### Часть 3

```python
x = np.arange(-5, 0, 0.001)
fig = plt.figure(figsize=(13, 8), dpi=120)
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim([-10,30])
axes.grid()
for a, b, c in task:
  y = [f(i, a, b) for i in x]
  axes.plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$') 
axes.plot(x, [0 for _ in x], 'm', label='f(x)=0')
axes.legend(loc=3, fontsize=12)
axes.set_xlabel('x',fontsize=12)
axes.set_ylabel('f(x)', fontsize=12)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных значениях $\alpha$ и $\beta$', fontsize=12)

x = np.arange(-100, -5, 0.001)
axes2 = fig.add_axes([0.2, 0.4, 0.6, 0.5])
axes2.set_ylim([0.8,1.1])
axes2.grid()
for a, b, c in task:
  y = [f(i, a, b) for i in x]
  axes2.plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$')
axes2.plot(x, [0 for _ in x], 'm', label='f(x)=0')
```

## Задача №2

```python
x = np.arange(-4, 4, 0.1)
task = [(1, 0, 'r'), (0, 0, 'g'), (0, 1, 'b'), (0, 2, 'm'), (1, -1, 'y'), (2, -1, 'c')]
fig = plt.figure(figsize=(13, 8), dpi=120)
axes = fig.add_axes([0, 0, 1, 1])
axes.grid()
for a, b, c in task:
  y = [f(i, a, b) for i in x]
  axes.plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$') 

axes.legend(loc=4, fontsize=12)
axes.set_xlabel('x',fontsize=12)
axes.set_ylabel('f(x)', fontsize=12)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных значениях $\alpha$ и $\beta$', fontsize=12)
```

Выводы:
- При $\alpha$=1 и $\beta$=0 функция равна 2 на всем промежутке (функция вырождается в $f(x)=\frac{1 + 1}{1}=2$)
- При $\alpha$=0 и $\beta$=0 функция равна 2 на всем промежутке (функция вырождается в $f(x)=\frac{1 + 1}{1}=2$)
- При $\alpha$=0 и $\beta$=1 функция равна 1 на всем промежутке (функция вырождается в $f(x)=\frac{x}{x}=1$)
- При $\alpha$=0 и $\beta$=2 функция равна 1 на всем промежутке (функция вырождается в $f(x)=\frac{x^2}{x^2}=1$)
- При $\alpha$=1 и $\beta$=-1 функция (функция $f(x)=\frac{x^-1 + 1}{x^-1}=1+\frac{1}{x^-1} = 1 + x$)
- При $\alpha$=2 и $\beta$=-1 функция (функция $f(x)=\frac{x^-1 + 2^-1}{x^-1}=1+\frac{0.5}{x^-1}=1+\frac{x}{2}$)
- Все полученные функции являются линейными

## Задача №3

```python
x = np.arange(0, 2, 0.001)
task = [(1, 0.5, 'r'), (1, -0.5, 'g'), (1, -1.5, 'b')]
fig = plt.figure(figsize=(13, 8), dpi=120)
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim([0,5])
axes.grid()
for a, b, c in task:
  y = [f(i, a, b) if i!= 0 else np.nan for i in x]
  axes.plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$') 

axes.legend(loc=0, fontsize=12)
axes.set_xlabel('x',fontsize=12)
axes.set_ylabel('f(x)', fontsize=12)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных значениях $\alpha$ и $\beta$', fontsize=12)
```

Выводы:
- При всех условиях входных данных функция не определена на отрицательной части оси абсцисс в виду того, что четная степенная функция корня не может иметь отрицательных аргументов
- Все три функции не определены при x=0
- При малых значениях х, устремленных к 0, значение первой функции асимптотически стремится к 0, а второй и третьей функция к 1
- Первая функция является убывающей на всем промежутке и вогнутой
- Вторая функция является возрастающей и вогнутой
- Третья функция является возрастающей и выпуклой

## Задача №4

```python
x = np.arange(0.1, 5, 0.1)
base = [(1, 0, 'b--'), (1, -1, 'r--')]
task = [[(1, 0.5, 'm'), (1, 0.8, 'g')], 
        [(1, -0.5, 'm'), (1, -0.8, 'g')], 
        [(1, -1.5, 'm'), (1, -2.5, 'g')]]
fig, axes = plt.subplots(1, 3, figsize=(13,8), dpi=120)
plt.suptitle(r'Графики функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных значениях $\alpha$ и $\beta$', fontsize=12)
for i in range(3):
    axes[i].set_ylim([0,5])
    axes[i].grid()
    for a, b, c in base:
        y = [f(i, a, b) for i in x]
        axes[i].plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$')
    for a, b, c in task[i]:
        y = [f(i, a, b) for i in x]
        axes[i].plot(x, y, c, label=f'$\\alpha={a}, \\beta={b}$')
    axes[i].legend(loc=0, fontsize=12)
    axes[i].set_xlabel('x',fontsize=12)
    axes[i].set_ylabel('f(x)', fontsize=12)
```

Выводы:
- Таким образом мы убедились, что при любых значениях $\alpha$ и $\beta$ график исследуемой функции проходит через точку (1, 2)
